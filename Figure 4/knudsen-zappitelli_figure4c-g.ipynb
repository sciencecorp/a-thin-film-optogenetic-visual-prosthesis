```python
# FIGURE 4d-e: example gamma responses to LED flash post sodium iodate:
%matplotlib inline
from IPython.display import set_matplotlib_formats
from matplotlib import cm
import matplotlib.pyplot as plt
import numpy as np
import pickle

# LOAD DATA:
#file = '/Volumes/Science/Biology/Paper_data_and_code/Figure 4/led_rabbit_data.pkl' # uncomment for d-e
file = '/Volumes/Science/Biology/Paper_data_and_code/Figure 4/lum_rabbit_data.pkl' # uncomment for f-g
with open(file, 'rb') as h:
    rbt = pickle.load(h)
    print(rbt.keys())
    
cond = rbt['condition'].astype(int)
if np.min(cond)==1:
    cond=cond-1

unique = np.unique(cond)
viridis = cm.get_cmap('viridis',len(unique))
col = viridis(range(len(unique)))
col[:,-1]=0.7

fig,ax=plt.subplots(1,2,layout='constrained')
fig.set_size_inches([6,3])

mm=[]
t=np.arange(-rbt['params']['pre'],rbt['params']['post'],1)

for u in unique:
    ind = np.where(cond==u)[0]
    m = np.mean(np.mean(rbt['gamma'][:,:,ind],axis=-1),axis=-1)

    ax[0].plot(t,m,lw=2,color=col[u-1,:],label=str(u))
    m = np.mean(rbt['gamma'][:,:,ind],axis=-1)
    mm.append(np.max(m[100:200,:],axis=0))
    #T=np.mean(np.max(rbt['gamma'][100:200,:,ind],axis=0),axis=1)

    #mm.append(T)
    
mm=np.array(mm)
print(mm.shape)
for i in range(mm.shape[0]):
    ax[1].plot(i*np.ones(mm.shape[-1]),mm[i,:],'o',color=col[i,:])
    ax[1].plot([i-0.5,i+0.5],[np.mean(mm[i,:]),np.mean(mm[i,:])],color=col[i,:],lw=2)

ax[0].set_ylabel("Average $\gamma$ power (mV^2)")
ax[0].set_xlabel('Time from flexled on (s)')
ax[0].plot([0,0],ax[0].get_ylim(),'r',lw=2)
ax[0].spines['top'].set_visible(False)
ax[0].spines['right'].set_visible(False)
ax[0].set_ylim([-4,23])

ax[1].spines['top'].set_visible(False)
ax[1].spines['right'].set_visible(False)
ax[1].set_ylabel("Mean peak power (mV^2)")
ax[1].set_xlabel('LED Power')

pwr = [2.2,4.1,10.6,17.3,23.1,29.3,35.1,41.1,46.9,52.8,58.7]
ax[1].set_xticks(range(mm.shape[0]))
ax[1].set_xticklabels(pwr)
ax[1].set_ylim([0,65])
```
