```python
# Suppplementary Figure 7: intravitreal injection of channel blockers reduces visual responses in V1
import matplotlib.pyplot as plt
import numpy as np
import scipy
import pickle

file = '/Volumes/Science/Biology/Paper_data_and_code/Supplemental Fig 7/blocker_data.pkl' 
with open(file, 'rb') as h:
    blk_data = pickle.load(h)

condit=10
i0 = np.where(blk_data['t0']['condition']==condit)[0]
m0 = np.mean(np.mean(blk_data['t0']['gamma'][:,:,i0],axis=1),axis=-1)
s0 = np.std(np.mean(blk_data['t0']['gamma'][:,:,i0],axis=1),axis=-1)/np.sqrt(len(i0))
i1 = np.where(blk_data['t1']['condition']==condit)[0]
m1 = np.mean(np.mean(blk_data['t1']['gamma'][:,:,i1],axis=1),axis=-1)
s1 = np.std(np.mean(blk_data['t1']['gamma'][:,:,i1],axis=1),axis=-1)/np.sqrt(len(i1))

t=np.arange(-100,400,1)

fig,ax = plt.subplots()
ax.fill_between(t,m0-s0,m0+s0,facecolor='tab:blue',edgecolor='none',alpha=0.4)
ax.fill_between(t,m1-s1,m1+s1,facecolor='tab:orange',edgecolor='none',alpha=0.4)

ax.plot(t,m0,lw=2.5,label='pre',color='tab:blue')
ax.plot(t,m1,lw=2.5,label='post',color='tab:orange')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.set_xlabel('Time from stimulus on (ms)')
ax.set_ylabel('mV$^{2}$')

yli=ax.get_ylim()
plt.plot([0,0],yli,'k--',lw=2)
plt.plot([200,200],yli,'k--',lw=2)

ax.legend(loc=2)
ax.set_title('$\gamma$ power, lum = max')
```
